#### Helping functions

```python
def reactions_to_df(reactions):
    rks_df = pd.DataFrame(columns=['reaction', 'reaction_name'])
    for rk in reactions:
        rks_df = rks_df.append({'reaction': rk.id, 'reaction_name': rk.name}, ignore_index=True)
    return rks_df

def get_filtered_prod_cons_table(model, model_name, condition, outdir, met, prop = 0.02):
    """returns the filtered production and consumption table of the metaboite specified"""
    sum = model.metabolites.get_by_id(met).summary()
    production = sum.producing_flux[sum.producing_flux.percent > prop]
    # multiply percent column and round two two digits
    production['percent'] = production['percent'].apply(lambda x: round(x*100, 2))
    production = production.sort_values(by='percent', ascending=False)
    consumption = sum.consuming_flux[sum.consuming_flux.percent > prop]
    consumption['percent'] = consumption['percent'].apply(lambda x: round(x*100, 2))
    consumption = consumption.sort_values(by='percent', ascending=False)
    # append model name to index column
    production.index = production.index + '_' + model_name + '_' + condition
    consumption.index = consumption.index + '_' + model_name + '_' + condition
    
    reaction_df = reactions_to_df(model.reactions)
    reaction_df['model'] = model_name
    # join production and consumption with reaction_df: left join on reaction
    production = pd.merge(production, reaction_df, on='reaction', how='left')
    consumption = pd.merge(consumption, reaction_df, on='reaction', how='left')

    # check if outdir exists
    if not os.path.exists(outdir):
        os.makedirs(outdir)
    production.to_csv(f'{outdir}{model_name}_{condition}_production.csv', index=False)
    consumption.to_csv(f'{outdir}{model_name}_{condition}_consumption.csv', index=False)
    return production, consumption
```

```python
# load all files in a directory with pattern *_atp_consumption.csv and concatenate them
def combine_all_files_per_condition(dir, condition):
    """returns a dataframe with all atp consumption files in a directory"""
    files_consum = [f for f in os.listdir(dir) if f.endswith(f'_{condition}_consumption.csv') and not f.startswith('all_')]
    files_prod = [f for f in os.listdir(dir) if f.endswith(f'_{condition}_production.csv') and not f.startswith('all_')]
    df_consum = pd.DataFrame()
    df_prod = pd.DataFrame()

    # error if production and consumption files are not equal
    if len(files_consum) != len(files_prod):
        raise ValueError('Number of production and consumption files are not equal.')

    for idx in range(len(files_consum)):
        df_consum = df_consum.append(pd.read_csv(dir + files_consum[idx]))
        df_prod = df_prod.append(pd.read_csv(dir + files_prod[idx]))
    # store these files
    df_consum.to_csv(f'{dir}all_{condition}_consumption.csv', index=False)
    df_prod.to_csv(f'{dir}all_{condition}_production.csv', index=False)
    return df_prod, df_consum

outdir = '../results/currency_metabolites/'
combine_all_files_per_condition(outdir, 'atp')
combine_all_files_per_condition(outdir, 'nadh')
combine_all_files_per_condition(outdir, 'nadph')
```
